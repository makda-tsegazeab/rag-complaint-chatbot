```python
# Task 1: Exploratory Data Analysis and Data Preprocessing
# Using the actual CFPB dataset: complaints.csv

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set visualization style
plt.style.use('ggplot')
sns.set_palette("husl")
%matplotlib inline

print("="*70)
print("INTELLIGENT COMPLAINT ANALYSIS FOR FINANCIAL SERVICES")
print("Task 1: Exploratory Data Analysis and Data Preprocessing")
print("="*70)

# 1. LOAD THE DATA
print("\n1. Loading CFPB complaint dataset...")
print("-"*50)

df = pd.read_csv('../data/raw/complaints.csv', low_memory=False)
print(f"✓ Dataset loaded successfully!")
print(f"  Shape: {df.shape[0]:,} rows × {df.shape[1]} columns")

# Display basic info
print("\nFirst 3 rows:")
display(df.head(3))

print("\nDataset info:")
print(df.info())

# 2. INITIAL DATA EXPLORATION
print("\n" + "="*70)
print("2. Initial Data Exploration")
print("="*70)

# Check column names
print("\nColumn Names:")
for i, col in enumerate(df.columns.tolist(), 1):
    print(f"{i:2}. {col}")

# Check for missing values
print("\nMissing values (top 15 columns):")
missing = df.isnull().sum().sort_values(ascending=False)
missing_percent = (missing / len(df)) * 100
missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_percent
})
print(missing_df.head(15))

# 3. IDENTIFY RELEVANT COLUMNS
print("\n" + "="*70)
print("3. Identifying Relevant Columns for Analysis")
print("="*70)

# Based on CFPB dataset structure, we expect these columns:
expected_columns = {
    'Product': 'Financial product type',
    'Consumer complaint narrative': 'The actual complaint text',
    'Issue': 'Main complaint category',
    'Sub-issue': 'Detailed complaint category',
    'Company': 'Company being complained about',
    'State': 'US state code',
    'Date received': 'Date complaint was received',
    'Complaint ID': 'Unique identifier'
}

print("\nExpected columns for our analysis:")
for col, desc in expected_columns.items():
    exists = "✓" if col in df.columns else "✗"
    print(f"  {exists} {col:35} - {desc}")

# Check which expected columns we have
available_cols = [col for col in expected_columns if col in df.columns]
print(f"\n✓ Found {len(available_cols)} of {len(expected_columns)} expected columns")

# 4. ANALYZE PRODUCT DISTRIBUTION
print("\n" + "="*70)
print("4. Analyzing Product Distribution")
print("="*70)

if 'Product' in df.columns:
    print(f"\nUnique products: {df['Product'].nunique()}")
    
    # Show all product categories
    product_counts = df['Product'].value_counts()
    print("\nProduct Distribution:")
    for product, count in product_counts.items():
        percentage = (count / len(df)) * 100
        print(f"  {product:40} - {count:7,} ({percentage:5.1f}%)")
    
    # Visualize
    plt.figure(figsize=(14, 8))
    top_products = product_counts.head(15)
    bars = plt.barh(range(len(top_products)), top_products.values)
    plt.yticks(range(len(top_products)), top_products.index)
    plt.xlabel('Number of Complaints')
    plt.title('Top 15 Financial Products by Complaint Volume')
    plt.gca().invert_yaxis()
    
    # Add value labels
    for i, bar in enumerate(bars):
        width = bar.get_width()
        plt.text(width + 100, bar.get_y() + bar.get_height()/2, 
                f'{width:,}', ha='left', va='center')
    
    plt.tight_layout()
    plt.savefig('product_distribution.png', dpi=100, bbox_inches='tight')
    plt.show()
    
else:
    print("✗ 'Product' column not found. Searching for similar columns...")
    for col in df.columns:
        if 'product' in col.lower():
            print(f"  Found potential: {col}")
            print(f"  Sample values: {df[col].unique()[:5]}")

# 5. FILTER TO REQUIRED PRODUCTS
print("\n" + "="*70)
print("5. Filtering to Required Product Categories")
print("="*70)

# Define the 4 product categories from project requirements
required_products_mapping = {
    'Credit Cards': [
        'Credit card', 'Credit card or prepaid card', 'Prepaid card'
    ],
    'Personal Loans': [
        'Personal loan', 'Payday loan', 'Vehicle loan or lease',
        'Student loan', 'Mortgage'
    ],
    'Savings Accounts': [
        'Savings account', 'Checking or savings account',
        'Bank account or service'
    ],
    'Money Transfers': [
        'Money transfer', 'Virtual currency', 'Money transfers'
    ]
}

if 'Product' in df.columns:
    # Create reverse mapping for filtering
    product_to_category = {}
    for category, products in required_products_mapping.items():
        for product in products:
            product_to_category[product] = category
    
    # Filter dataframe to only include our required products
    required_product_list = []
    for products in required_products_mapping.values():
        required_product_list.extend(products)
    
    df_filtered = df[df['Product'].isin(required_product_list)].copy()
    
    # Add product_category column
    df_filtered['product_category'] = df_filtered['Product'].map(product_to_category)
    
    print(f"\nFiltering results:")
    print(f"  Original dataset: {len(df):,} complaints")
    print(f"  After filtering:  {len(df_filtered):,} complaints")
    print(f"  Removed:          {len(df) - len(df_filtered):,} complaints")
    
    # Show distribution of our 4 categories
    print("\nDistribution across 4 required categories:")
    category_counts = df_filtered['product_category'].value_counts()
    for category, count in category_counts.items():
        percentage = (count / len(df_filtered)) * 100
        print(f"  {category:20} - {count:7,} ({percentage:5.1f}%)")
    
    # Visualize filtered distribution
    plt.figure(figsize=(10, 6))
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']
    wedges, texts, autotexts = plt.pie(
        category_counts.values, 
        labels=category_counts.index,
        colors=colors,
        autopct='%1.1f%%',
        startangle=90
    )
    plt.title('Complaint Distribution Across Financial Products\n(Filtered to 4 Categories)')
    plt.axis('equal')
    plt.tight_layout()
    plt.savefig('filtered_product_distribution.png', dpi=100, bbox_inches='tight')
    plt.show()
    
else:
    print("✗ Cannot filter products: 'Product' column not available")
    df_filtered = df.copy()

# 6. ANALYZE COMPLAINT NARRATIVES
print("\n" + "="*70)
print("6. Analyzing Consumer Complaint Narratives")
print("="*70)

narrative_col = 'Consumer complaint narrative'

if narrative_col in df_filtered.columns:
    # Analyze narrative availability
    has_narrative = df_filtered[narrative_col].notna()
    narratives_count = has_narrative.sum()
    
    print(f"\nNarrative Analysis:")
    print(f"  Total complaints: {len(df_filtered):,}")
    print(f"  With narratives:  {narratives_count:,} ({has_narrative.mean()*100:.1f}%)")
    print(f"  Without narratives: {(~has_narrative).sum():,} ({(~has_narrative).mean()*100:.1f}%)")
    
    # Filter to only complaints with narratives
    df_with_narratives = df_filtered[has_narrative].copy()
    
    # Calculate narrative lengths
    df_with_narratives['narrative_length'] = df_with_narratives[narrative_col].apply(
        lambda x: len(str(x).split()) if pd.notna(x) else 0
    )
    
    # Analyze length distribution
    print(f"\nNarrative Length Statistics:")
    print(f"  Mean length:    {df_with_narratives['narrative_length'].mean():.1f} words")
    print(f"  Median length:  {df_with_narratives['narrative_length'].median():.1f} words")
    print(f"  Std deviation:  {df_with_narratives['narrative_length'].std():.1f} words")
    print(f"  Min length:     {df_with_narratives['narrative_length'].min()} words")
    print(f"  Max length:     {df_with_narratives['narrative_length'].max()} words")
    
    # Visualize narrative lengths
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Histogram
    axes[0].hist(df_with_narratives['narrative_length'], bins=50, 
                 edgecolor='black', alpha=0.7, color='skyblue')
    axes[0].axvline(df_with_narratives['narrative_length'].mean(), 
                    color='red', linestyle='--', label=f'Mean: {df_with_narratives["narrative_length"].mean():.1f}')
    axes[0].axvline(df_with_narratives['narrative_length'].median(), 
                    color='green', linestyle='--', label=f'Median: {df_with_narratives["narrative_length"].median():.1f}')
    axes[0].set_xlabel('Word Count')
    axes[0].set_ylabel('Frequency')
    axes[0].set_title('Distribution of Complaint Narrative Lengths')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    
    # Box plot by product category
    if 'product_category' in df_with_narratives.columns:
        df_with_narratives.boxplot(
            column='narrative_length', 
            by='product_category',
            ax=axes[1],
            grid=False
        )
        axes[1].set_title('Narrative Length by Product Category')
        axes[1].set_ylabel('Word Count')
        axes[1].set_xlabel('Product Category')
        axes[1].tick_params(axis='x', rotation=45)
    else:
        axes[1].boxplot(df_with_narratives['narrative_length'])
        axes[1].set_title('Narrative Length Box Plot')
        axes[1].set_ylabel('Word Count')
    
    plt.suptitle('Analysis of Complaint Narrative Lengths', fontsize=14)
    plt.tight_layout()
    plt.savefig('narrative_length_analysis.png', dpi=100, bbox_inches='tight')
    plt.show()
    
    # Show examples
    print("\nSample Narratives by Length:")
    print("-" * 80)
    
    # Short narrative example
    short_idx = df_with_narratives['narrative_length'].idxmin()
    short_complaint = df_with_narratives.loc[short_idx]
    print(f"\nSHORTEST ({short_complaint['narrative_length']} words):")
    print(f"Product: {short_complaint.get('Product', 'N/A')}")
    print(f"Category: {short_complaint.get('product_category', 'N/A')}")
    print(f"Narrative: {short_complaint[narrative_col][:200]}...")
    
    # Long narrative example
    long_idx = df_with_narratives['narrative_length'].idxmax()
    long_complaint = df_with_narratives.loc[long_idx]
    print(f"\nLONGEST ({long_complaint['narrative_length']} words):")
    print(f"Product: {long_complaint.get('Product', 'N/A')}")
    print(f"Category: {long_complaint.get('product_category', 'N/A')}")
    print(f"Narrative: {long_complaint[narrative_col][:200]}...")
    
    # Average narrative example
    avg_length = df_with_narratives['narrative_length'].mean()
    closest_idx = (df_with_narratives['narrative_length'] - avg_length).abs().idxmin()
    avg_complaint = df_with_narratives.loc[closest_idx]
    print(f"\nAVERAGE LENGTH ({avg_complaint['narrative_length']} words):")
    print(f"Product: {avg_complaint.get('Product', 'N/A')}")
    print(f"Category: {avg_complaint.get('product_category', 'N/A')}")
    print(f"Narrative: {avg_complaint[narrative_col][:200]}...")
    
else:
    print(f"✗ Narrative column '{narrative_col}' not found!")
    print("Available text columns:")
    text_cols = [col for col in df_filtered.columns if any(word in col.lower() 
                for word in ['text', 'narrative', 'description', 'complaint', 'detail'])]
    for col in text_cols:
        print(f"  - {col}")

# 7. TEXT CLEANING AND PREPROCESSING
print("\n" + "="*70)
print("7. Text Cleaning and Preprocessing")
print("="*70)

import re
import string

def clean_complaint_text(text):
    """
    Clean complaint narratives for better embedding quality
    """
    if pd.isna(text):
        return ""
    
    text = str(text)
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove common boilerplate phrases (common in CFPB complaints)
    boilerplate_phrases = [
        r'i am writing to file a complaint',
        r'this is a complaint regarding',
        r'i would like to complain about',
        r'dear sir/madam',
        r'to whom it may concern',
        r'please be advised that',
        r'company name\s*:.*',
        r'account number\s*:.*',
        r'xxxx',
        r'xxx'
    ]
    
    for phrase in boilerplate_phrases:
        text = re.sub(phrase, '', text, flags=re.IGNORECASE)
    
    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    
    # Remove email addresses
    text = re.sub(r'\S+@\S+', '', text)
    
    # Remove phone numbers
    text = re.sub(r'\(?\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}', '', text)
    
    # Remove special characters but keep basic punctuation
    text = re.sub(r'[^\w\s.,!?;:\'\"()-]', ' ', text)
    
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text)
    
    # Remove leading/trailing whitespace
    text = text.strip()
    
    return text

if narrative_col in df_with_narratives.columns:
    print("\nApplying text cleaning to narratives...")
    
    # Apply cleaning
    df_with_narratives['cleaned_narrative'] = df_with_narratives[narrative_col].apply(
        clean_complaint_text
    )
    
    # Calculate cleaned lengths
    df_with_narratives['cleaned_length'] = df_with_narratives['cleaned_narrative'].apply(
        lambda x: len(x.split())
    )
    
    # Show cleaning results
    print("\nText Cleaning Results:")
    print(f"  Average words removed: {df_with_narratives['narrative_length'].mean() - df_with_narratives['cleaned_length'].mean():.1f}")
    
    # Show before/after examples
    sample_idx = df_with_narratives.index[0]
    print(f"\nExample - Before Cleaning ({df_with_narratives.loc[sample_idx, 'narrative_length']} words):")
    print("-" * 80)
    print(df_with_narratives.loc[sample_idx, narrative_col][:300])
    
    print(f"\nExample - After Cleaning ({df_with_narratives.loc[sample_idx, 'cleaned_length']} words):")
    print("-" * 80)
    print(df_with_narratives.loc[sample_idx, 'cleaned_narrative'][:300])
    
else:
    print("✗ Cannot clean text: narrative column not available")

# 8. ADDITIONAL ANALYSES
print("\n" + "="*70)
print("8. Additional Analyses")
print("="*70)

# Analyze issues by product category
if 'Issue' in df_with_narratives.columns and 'product_category' in df_with_narratives.columns:
    print("\nTop Issues by Product Category:")
    print("-" * 80)
    
    for category in df_with_narratives['product_category'].unique():
        category_df = df_with_narratives[df_with_narratives['product_category'] == category]
        top_issues = category_df['Issue'].value_counts().head(5)
        
        print(f"\n{category}:")
        for issue, count in top_issues.items():
            percentage = (count / len(category_df)) * 100
            print(f"  • {issue:40} - {count:5} ({percentage:4.1f}%)")

# Analyze complaints over time
if 'Date received' in df_with_narratives.columns:
    print("\nAnalyzing Complaint Trends Over Time...")
    
    # Convert to datetime
    df_with_narratives['date_received'] = pd.to_datetime(
        df_with_narratives['Date received'], errors='coerce'
    )
    
    # Extract year-month for trend analysis
    df_with_narratives['year_month'] = df_with_narratives['date_received'].dt.to_period('M')
    
    # Plot monthly complaint volume
    monthly_counts = df_with_narratives['year_month'].value_counts().sort_index()
    
    plt.figure(figsize=(14, 6))
    monthly_counts.plot(kind='line', marker='o', linewidth=2)
    plt.title('Monthly Complaint Volume Trend')
    plt.xlabel('Month')
    plt.ylabel('Number of Complaints')
    plt.grid(True, alpha=0.3)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig('monthly_complaint_trend.png', dpi=100, bbox_inches='tight')
    plt.show()
    
    print(f"Date range: {df_with_narratives['date_received'].min().date()} to {df_with_narratives['date_received'].max().date()}")
    print(f"Total months of data: {len(monthly_counts)}")
    print(f"Average monthly complaints: {monthly_counts.mean():.0f}")

# 9. SAVE PROCESSED DATA
print("\n" + "="*70)
print("9. Saving Processed Data")
print("="*70)

# Prepare final dataframe for saving
final_columns = [
    'Complaint ID', 'date_received', 'product_category', 'Product',
    'Issue', 'Sub-issue', 'Company', 'State',
    'cleaned_narrative', 'narrative_length', 'cleaned_length'
]

# Filter to available columns
available_final_cols = [col for col in final_columns 
                       if col in df_with_narratives.columns or 
                       (col == 'Complaint ID' and 'Complaint ID' in df.columns)]

final_df = df_with_narratives[available_final_cols].copy()

# Rename columns for consistency
column_rename = {
    'Complaint ID': 'complaint_id',
    'Product': 'original_product',
    'Issue': 'issue',
    'Sub-issue': 'sub_issue',
    'Company': 'company',
    'State': 'state',
    'cleaned_narrative': 'consumer_complaint_narrative',
    'date_received': 'date_received'
}

final_df = final_df.rename(columns=column_rename)

# Save to CSV
output_path = '../data/processed/filtered_complaints.csv'
final_df.to_csv(output_path, index=False)

print(f"\n✓ Processed data saved to: {output_path}")
print(f"✓ Total complaints saved: {len(final_df):,}")
print(f"✓ Columns saved: {len(final_df.columns)}")
print(f"\nFinal columns:")
for col in final_df.columns:
    print(f"  • {col}")

print(f"\nFirst 3 rows of saved data:")
display(final_df.head(3))

# 10. SUMMARY REPORT
print("\n" + "="*70)
print("10. TASK 1 SUMMARY REPORT")
print("="*70)

print("\n📊 DATA OVERVIEW:")
print(f"  • Raw dataset: {len(df):,} total complaints")
print(f"  • After filtering to 4 product categories: {len(df_filtered):,} complaints")
print(f"  • With narratives: {len(df_with_narratives):,} complaints")
print(f"  • Final processed dataset: {len(final_df):,} complaints")

print("\n📈 PRODUCT DISTRIBUTION:")
if 'product_category' in final_df.columns:
    for category, count in final_df['product_category'].value_counts().items():
        percentage = (count / len(final_df)) * 100
        print(f"  • {category:20} - {count:6,} complaints ({percentage:5.1f}%)")

print("\n📝 NARRATIVE STATISTICS:")
if 'narrative_length' in final_df.columns:
    print(f"  • Average length: {final_df['narrative_length'].mean():.1f} words")
    print(f"  • Median length:  {final_df['narrative_length'].median():.1f} words")
    print(f"  • Range: {final_df['narrative_length'].min()} to {final_df['narrative_length'].max()} words")

print("\n📅 TEMPORAL COVERAGE:")
if 'date_received' in final_df.columns:
    date_range = final_df['date_received'].dropna()
    if len(date_range) > 0:
        print(f"  • From: {date_range.min().date()}")
        print(f"  • To:   {date_range.max().date()}")

print("\n💾 OUTPUT FILES:")
print(f"  • Processed data: data/processed/filtered_complaints.csv")
print(f"  • Visualizations: product_distribution.png")
print(f"  • Visualizations: filtered_product_distribution.png")
print(f"  • Visualizations: narrative_length_analysis.png")
print(f"  • Visualizations: monthly_complaint_trend.png")

print("\n" + "="*70)
print("✅ TASK 1 COMPLETED SUCCESSFULLY!")
print("="*70)
print("\nNext Steps:")
print("1. Review the generated visualizations")
print("2. Check the processed data in 'data/processed/filtered_complaints.csv'")
print("3. Commit your changes: git add . && git commit -m 'Completed Task 1: EDA and preprocessing'")
print("4. Merge to dev branch: git checkout dev && git merge eda-preprocessing")
print("5. Proceed to Task 2: Text Chunking and Embedding")
```
